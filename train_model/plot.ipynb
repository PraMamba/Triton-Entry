```python
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
data1 = []
with open('logs/qwen2-0.5B-with-kernel/log_history.json', 'r') as f:
    for line in f.readlines():
        try:
            data1.append(json.loads(line))
        except:
            pass
info1 = data1.pop(-1)
df1 = pd.DataFrame(data1)
data2 = []
with open('logs/qwen2-0.5B-without-kernel/log_history.json', 'r') as f:
    for line in f.readlines():
        try:
            data2.append(json.loads(line))
        except:
            pass
info2 = data2.pop(-1)
df2 = pd.DataFrame(data2)
```

```python
start = 100
plt.figure(figsize=(10,6))
plt.plot(df1['loss'][start:], alpha=0.5)
plt.plot(df2['loss'][start:], alpha=0.5)
plt.legend(['with-kernel', 'without-kernel'], fontsize=20)
plt.xlabel('steps', size=15)
plt.ylabel('training loss', size=15)
plt.show()
```
