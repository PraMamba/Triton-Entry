```python
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
def get_df(path, return_info=False):
    data = []
    with open(path, 'r') as f:
        for line in f.readlines():
            # data.append(json.loads(line))
            try:
                data.append(json.loads(line))
            except:
                pass
    # print(len(data))
    data.pop(-1)
    df = pd.DataFrame(data)
    if return_info: return df,info
    return df
```

# Qwen2.5-0.5B-pretrain 

```python
df1 = get_df('logs/qwen2-0.5B-with-kernel/log_history.json')
df2 = get_df('logs/qwen2-0.5B-without-kernel/log_history.json',)
df3 = get_df('logs/qwen2-0.5B-with-kernel-unsloth-loss-pretrain/log_history.json')
df4 = get_df('logs/qwen2-0.5B-with-kernel-pretrain-bf16-loss/log_history.json')
```

```python
start = 100
plt.figure(figsize=(10,6))
plt.plot(df1['loss'][start:], alpha=0.5)
plt.plot(df2['loss'][start:], alpha=0.5)
plt.plot(df3['loss'][start:], alpha=0.5)
plt.plot(df4['loss'][start:], alpha=0.5)
plt.legend(['with-kernel', 'without-kernel', 'kernel+unsloth', 'kernel+bf16 loss'], fontsize=20)
plt.xlabel('steps', size=15)
plt.ylabel('training loss', size=15)
plt.show()
```
